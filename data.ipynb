```python
import pandas as pd
```

```python
df = pd.read_csv('Indian_cities.csv')
df
```

```python
cities = df['name_of_city']
cities =pd.DataFrame(cities)
cities
cities = cities.rename(columns = {'name_of_city' : 'city'})
cities
```

```python
import json
cityData = cities.to_json(orient= 'records')
# cityData1 = cities.to_csv()
cityData
```

```python
json_object = json.dumps(cityData, indent = 2)
# js = json.loads(json_object.replace("\\"," "))
js =json_object.replace("\\","")
js
```

```python

# Writing to sample.json
with open("Cities.json", "w") as file:
    file.write(js)
```
